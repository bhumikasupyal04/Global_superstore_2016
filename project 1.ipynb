-  Observe the columns 'Order Date' and 'Ship Date' :-

```python

df['Order Date'] = df['Order Date'].str.replace('/','-')
df['Ship Date'] =df ['Ship Date'].str.replace('/','-')

print(df[['Order Date','Ship Date']].head(10))
```

-  Observe the columns 'Sales' and 'Profit' : 

```python

df['Sales'] =df['Sales'].str.replace('$','')
df['Sales'] =df['Sales'].str.replace(',','')
df['Profit'] =df['Profit'].str.replace('$','')
df['Profit'] =df['Profit'].str.replace(')','')
df['Profit'] =df['Profit'].str.replace('(','')
df['Profit'] =df['Profit'].str.replace(',','')
```

```python
df = df.drop(columns=['Postal Code'])
```

```python
df['Sales']=df['Sales'].astype(float)
df['Profit']=df['Profit'].astype(float)
df['Shipping Cost']=df['Shipping Cost'].astype(float)
df['Order Date']=df['Order Date'].astype('datetime64[ns]')
df['Ship Date']=df['Ship Date'].astype('datetime64[ns]')
```

- 8. Find the overall rank of "India" where the 'Profit' is maximum under the category 'Technology'.

```python
import pandas as pd

df_tech = df[df['Category'] == 'Technology']

country_profit = df_tech.groupby('Country')['Profit'].sum().reset_index()

country_profit['Rank'] = country_profit['Profit'].rank(method='dense', ascending=False)

country_profit = country_profit.sort_values('Rank')

india_rank = country_profit[country_profit['Country'] == 'India']['Rank'].values[0]

print(f"Overall rank of India (by total Profit under Technology): {int(india_rank)}")
```

- 9. Display the data with min, max, average and std of 'Profit' & 'Sales' for each Sub-Category under each Category

```python
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
df['Sales'] = pd.to_numeric(df['Sales'],errors='coerce')

grouped = df.groupby(['Category', 'Sub-Category']).agg({
    'Profit': ['min', 'max', 'mean', 'std'],

    'Sales': ['min', 'max', 'mean', 'std']
})

print(grouped)
```
